# Q1.Create a function in python to read the text file and replace specific content of the file.

When filename is given it reds the file by line by line and replace the string(target).

```python
def find_replace(x):
    s1 = "placement" 
    r1 = "screening"
    with open( x , 'r') as file:
        data = file.read()
        data = data.replace(s1, r1)
    with open(x , 'w') as file:
        file.write(data)
```

# Q2. Abstract Class is class in which we make bluprint or methoods to use it in child class.

```python
#First we import module abc and from that we import ABC (Abstract Base Class) and abstractmethod
```

```python
from abc import ABC, abstractmethod
```

```python
#Then we create bstract class and here we define method as abstract method by putting @abstractmethod
```

```python
class Professional(ABC):
    @abstractmethod
    def fun1(self):
        pass
```

```python
#Now we create derived class from base class(abstract base class) and use the function of base class and override it.
```

```python
class Doctor(Professional):
    def fun1(self):
        print("I am", __class__.__name__)

    def fun2(self):
        print("This is function 2")


class Engineer(Professional):
    def fun1(self):
        print("I am", __class__.__name__)


class Teacher(Professional):
    def fun1(self):
        print("I am", __class__.__name__)
```

```python
d = Doctor()
d.fun1()
d.fun2()
e = Engineer()
e.fun1()
t = Teacher()
t.fun1()
```

# Multiple Inheritance is type of inheritance in which there two paernt class and one child class.

```python
# First we crete one prent class
```

```python
class Base1(object): 
    #print('Base1')
    def __init__(self): 
        self.str1 = "Geek1"
        print("Base1") 
        
    def temp(self):
        print('temp1')
```

```python
# Second we create another parent class.
```

```python
class Base2(object): 
    #print('Base2')
    def __init__(self): 
        self.str2 = "Geek2"        
        print("Base2") 
        
    def temp(self):
        print('temp2')
```

```python
# THird we create child class and call other parent classes.
```

```python
class Derived(Base2, Base1): 
    print('Derived')
    def __init__(self): 
          
        # Calling constructors of Base1 
        # and Base2 classes 
        Base1.__init__(self) 
        Base2.__init__(self) 
        print("Derived") 
          
    def printStrs(self): 
        print(self.str1, self.str2) 
```

```python
ob1 = Derived()
print("===========")
ob1.temp()
```
